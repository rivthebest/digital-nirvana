#### compare_models() is the second step and the very important step where we select a particular model which can be used as the optimized score grid values of MAE, MSE, RMSE, R2, RMSLE, MAPE are presented across the respective model.
#### The method is using blacklist parameters which will prevent assessment for the above mentioned statistical score for the respective models. Generally those models are skipped whose training time is high. Model abbreviation code is used for model name selection. Out of 25 6 models are skipped from evaluation as those models take more training time.
#### K-fold cross validation is also used where fold value is 5
#### According to the problem score is sorted with respect to RMSE.
#### This activity requires 2-3 hours approx or more depending on the dataset and set of models chosen for the score grid.

```python
compare_models(blacklist=['tr','ransac','svm'], fold=5, sort='RMSE', turbo=True)
```

#### 3rd Step is create_model() which helps to create a speciic model passed as string parameter and the respective score grid along with mean and SD are calculated with a 10-old cross validation

#### Creating CatBoost Model

```python
catboost = create_model('catboost')
```

#### Tuning CatBoost Model

```python
tuned_catboost = tune_model(catboost, optimize='RMSE')
```

#### After creating the model tuning_model() method is called and same statistical score are evaluated along with mean and SD with a 10-Fold Cross Validation. The method takes create_model() object as parameter

#### Creating LightGBM Model

```python
lightgbm = create_model('lightgbm')
```

#### Tuning LightGBM Model

```python
tuned_lightgbm = tune_model(lightgbm,optimize='RMSE')
```

#### interpret_model() is used visualize performance of the created model as graphical mode. catboost model is one of the models which use interpret_model() as visualization. The other 2 methods are evaluate_model() and plot_model() does not work for catboost model.

```python
interpret_model(tuned_catboost)
```
